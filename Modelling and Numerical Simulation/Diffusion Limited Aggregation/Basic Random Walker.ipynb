```python
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline

plt.rcParams["font.family"] = "monospace" #sets the fonts to be used by the plots later on
plt.rcParams.update({'font.size': 18})
```

```python
def random_walk(N): 
    
    '''The function starts with empty arrays for x and y. Starting from 0 the walker takes a random step
    (characterised by a random value of -1,0,1). These random steps are "collected" in the respective empty 
    x and y arrays arrays so they can be plotted'''
    
    xpath=np.zeros(N) #create empty arrays 
    ypath=np.zeros(N)
    for i in range(1,N): #This range always leaves the first position as the origin 
        xpath[i]=np.random.randint(-1,2)+xpath[i-1]
        ypath[i]=np.random.randint(-1,2)+ypath[i-1]
    

    return xpath,ypath



def display_walk(N):
    
    '''Creates a random walker pattern and plots it with 
       matplotlib'''
    
    xwalk, ywalk = random_walk(N)

    plt.figure(figsize=(10,10))
    plt.plot(0,0,'ro') #mark the seed with red spot 
    plt.plot(xwalk,ywalk,'b-')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

```

```python
display_walk(10000)
```
