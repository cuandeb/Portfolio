```python
#import the required packages/modules
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
```

The Julia set is an important concept in fractal theory. Given a complex number $a$, a point $z$ in the complex plane is said to be in the filled-in Julia set of a function 

$$f(z) = zÂ² + a$$

if the iteration of the function over the point does not finish with the point diverging i.e. going to infinity. It can be proved that if at some iterate of a point under f(z) the result has a modulus larger than 2 and larger than the modulus of a, this point will finish going to infinity.

In this notebook we build and plot the filled-in Julia sets for a series of complex numbers with real and imaginary coefficients in the interval of $-1 < Re(z)$, $Im(z) < 1$. The values below are arbitrarily chosen mainly for the intricate patterns they produce . It is taken that the point belongs to the Julia set once the previous conditions has not been fulfilled after $N_{iter} = 100$. 

```python
def julia(z,a,max_iter): 
    
    '''z = some complex number that will be iterated
       a = a constant complex number
       max_iter = max number of iterations
       
       The function iterates in the specified range unless it fulfills the specification: |z|>2
       at which point the loop breaks and the iteration number is returned'''    
    
    for it in range(max_iter): 
        z=(z**2)+a
        #print(z)
        if abs(z)>2: 
            break            
    return it 


def build_set(a, b):
    
    '''Returns array of the filled in Julia set constructed from an 
    input complex number of the form a+bi. Inputs a and b are floats.'''
    
    x_values=np.linspace(-1,1,1000) #our real and imaginary componants are between -1 and 1
    y_values=np.linspace(-1,1,1000)

    xlen=len(x_values) #creating a space of comparable size for our julia set
    ylen=len(y_values)

    c=complex(a, b) #value of 'a' that can be changed 

    visuals=np.zeros([xlen,ylen]) #create an empty array to fill

    for i in range(xlen): 
        for j in range(ylen): 
            zx=x_values[i]
            zy=y_values[j]
            z=complex(zx, zy)
            visuals[i,j]=julia(z,c,100)
    return visuals


def plot_julia_set(re, im):
    
    '''Constructs and plots the filled in Julia set of an 
       input complex number '''
    
    j_set = build_set(re,im)
    plt.figure(figsize=(18,18))
    plt.title("Julia set for $a=%s+%si$" %(re,im))
    plt.imshow(j_set,interpolation="nearest",cmap='magma')
    plt.show()
    #we can change the cmap but I think magma looks quite striking 
    

```

## Plotted Sets 

Now with our functions defined we can plot the Julia set for any complex numbers that fulfill the previous criteria 


```python
plot_julia_set(-0.7269,0.1889)
```
