```python
import extraction_utils as extract
import curvefit_utils as fit
import efficiency as eff
import calibration_resolution as calib
import datetime
import numpy as np

from collections import namedtuple

#directories containing raw spectrum data
CAESIUM_HPGE='HPGe/Caesium'
COBALT_HPGE='HPGe/Cobalt'
AMERICIUM_HPGE='HPGe/Americium'
BARIUM_HPGE='HPGe/Barium'
BACKGROUND_HPGE='HPGe/Background/background_45000s.Spe'

#csv file containing fitted peak data
TOTAL_HPGE="Fitted_Data/HPGe/total_hpge.csv"

#region of interest for each peak
GE_ROI={'cs':(2580,2660),
        'mid_ba':(300,340),
         'ba1':(1050,1250),
         'ba2':(1350,1600),
         'co1':(4575,4775),
         'co2':(5200,5400),                 
         'am2':(215,260),
         'ka':(80,175)   
         }

#date and time for original source actvitiy
UPSTAIRS_SOURCES=datetime.datetime(1979,2,1,12)
DOWNSTAIRS_SOURCES=datetime.datetime(1979,12,1,12)

Model = namedtuple('Model', ['name','model', 'estimates',
                             'params', 'components'])


#function parameters for each fit
GAUSSIAN_PARAMS = ('mu', 'sig', 'A')
GAUSSIAN_LINE_PARAMS = ('mu','sig','A','Slope','Intercept')
DOUBLE_LINE_PARAMS= ('mu1','sig1','A1',
                     'mu2','sig2','A2',
                     'Slope','Intercept')
DOUBLE_PEAK_PARAMS= ('mu1','sig1','A1',
                     'mu2','sig2','A2')


GAUSSIAN_MODEL = Model(
    name      = 'gaussian',
    model     = fit.gaussian,
    estimates = fit.gaussian_initial_estimates,
    params    = GAUSSIAN_PARAMS,
    components = fit.gaussian
)


GAUSSIAN_LINE_MODEL = Model(
    name      = 'gaussian line',
    model     = fit.gaussian_line,
    estimates = fit.gaussian_line_initial_estimates,
    params    = GAUSSIAN_LINE_PARAMS,
    components  = fit.gaussian_line_components
)

DOUBLE_GAUSSIAN_MODEL = Model(
    name      = 'double line',
    model     = fit.double_line,
    estimates = fit.double_line_initial_estimates,
    params    = DOUBLE_LINE_PARAMS,
    components = fit.double_gaussian_components
)


TWO_GAUSSIAN_MODEL = Model(
    name      = 'double peak',
    model     = fit.double_peak,
    estimates = fit.double_peak_initial_estimates,
    params    = DOUBLE_PEAK_PARAMS,
    components = fit.double_peak_components
)

#################################################

def hpge_spectra():
    caesium_list=fit.plot_content(CAESIUM_HPGE, BACKGROUND_HPGE, UPSTAIRS_SOURCES)        
    cobalt_list=fit.plot_content(COBALT_HPGE, BACKGROUND_HPGE, UPSTAIRS_SOURCES)
    barium_list=fit.plot_content(BARIUM_HPGE, BACKGROUND_HPGE, UPSTAIRS_SOURCES)
    americium_list=fit.plot_content(AMERICIUM_HPGE, BACKGROUND_HPGE, UPSTAIRS_SOURCES)
    fit.everything(caesium_list,GE_ROI['cs'],'cs',GAUSSIAN_LINE_MODEL,'HPGe')
    fit.everything(barium_list,GE_ROI['mid_ba'],'mid_ba',GAUSSIAN_LINE_MODEL,'HPGe')
    fit.everything(barium_list,GE_ROI['ba1'],'ba1',DOUBLE_GAUSSIAN_MODEL,'HPGe')
    fit.everything(barium_list,GE_ROI['ba2'],'ba2',TWO_GAUSSIAN_MODEL,'HPGe')
    fit.everything(americium_list,GE_ROI['am2'],'am2',GAUSSIAN_LINE_MODEL,'HPGe')
    fit.everything(cobalt_list,GE_ROI['co1'],'co1',GAUSSIAN_MODEL,'HPGe')
    fit.everything(cobalt_list,GE_ROI['co2'],'co2',GAUSSIAN_MODEL,'HPGe')
    
    
if __name__ == "__main__":
    hpge_spectra()
    calib.total_calibration_spec(HPGE_PEAKS)
    eff.final_source_efficiencies('HPGe')    
```
