```python
import extraction_utils as extract
import curvefit_utils as fit
import efficiency as eff
import calibration_resolution as calib
import datetime
import numpy as np

from collections import namedtuple

#directories containing raw spectrum data
CAESIUM_BGO='BGO/Caesium'
AMERICIUM_BGO='BGO/Americium'
COBALT_BGO='BGO/Cobalt'
BARIUM_BGO='BGO/Barium'
BACKGROUND_BGO='BGO/Background/background_1000s.Spe'

#csv file containing fitted peak data
BGO_PEAKS="Fitted_Data/BGO/total_bgo.csv"

#region of interest for each peak
BGO_ROI={'cs':(320,550),
         'mid_ba':(30,70),
         'ba':(125,300),
         'co':(610,950),
         'am':(15,60),
         'ka':(10,40)
         }

#date and time for original source actvitiy
UPSTAIRS_SOURCES=datetime.datetime(1979,2,1,12)
DOWNSTAIRS_SOURCES=datetime.datetime(1979,12,1,12)

Model = namedtuple('Model', ['name','model', 'estimates',
                             'params', 'components'])


#function parameters for each fit
GAUSSIAN_PARAMS = ('mu', 'sig', 'A')
GAUSSIAN_LINE_PARAMS = ('mu','sig','A','Slope','Intercept')
DOUBLE_LINE_PARAMS= ('mu1','sig1','A1',
                     'mu2','sig2','A2',
                     'Slope','Intercept')


GAUSSIAN_MODEL = Model(
    name      = 'gaussian',
    model     = fit.gaussian,
    estimates = fit.gaussian_initial_estimates,
    params    = GAUSSIAN_PARAMS,
    components = fit.gaussian
)


GAUSSIAN_LINE_MODEL = Model(
    name      = 'gaussian line',
    model     = fit.gaussian_line,
    estimates = fit.gaussian_line_initial_estimates,
    params    = GAUSSIAN_LINE_PARAMS,
    components  = fit.gaussian_line_components
)

DOUBLE_GAUSSIAN_MODEL = Model(
    name      = 'double line',
    model     = fit.double_line,
    estimates = fit.double_line_initial_estimates,
    params    = DOUBLE_LINE_PARAMS,
    components = fit.double_gaussian_components
)


#################################################

        



def bgo_spectra():
    caesium_list=fit.plot_content(CAESIUM_BGO, BACKGROUND_BGO,DOWNSTAIRS_SOURCES)        
    cobalt_list=fit.plot_content(COBALT_BGO, BACKGROUND_BGO,DOWNSTAIRS_SOURCES)
    barium_list=fit.plot_content(BARIUM_BGO, BACKGROUND_BGO,DOWNSTAIRS_SOURCES)
    americium_list=fit.plot_content(AMERICIUM_BGO, BACKGROUND_BGO,DOWNSTAIRS_SOURCES)
    fit.everything(barium_list, BGO_ROI['mid_ba'],'mid_ba',GAUSSIAN_LINE_MODEL,'BGO')
    fit.everything(barium_list, BGO_ROI['ba'],'ba',DOUBLE_GAUSSIAN_MODEL,'BGO')
    fit.everything(caesium_list, BGO_ROI['cs'],'cs',GAUSSIAN_LINE_MODEL,'BGO')    
    fit.everything(americium_list, BGO_ROI['am'],'am2',GAUSSIAN_LINE_MODEL,'BGO')
    fit.everything(cobalt_list, BGO_ROI['co'],'co',DOUBLE_GAUSSIAN_MODEL,'BGO')


    

      
    
if __name__ == "__main__":
    bgo_spectra()
    calib.total_calibration_spec(BGO_PEAKS)
    eff.final_source_efficiencies('BGO')
```
