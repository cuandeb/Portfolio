```python
import extraction_utils as extract
import curvefit_utils as fit
import efficiency as eff
import calibration_resolution as calib
import datetime
import numpy as np

from collections import namedtuple

#directories containing raw spectrum data
CAESIUM_NAI='NaI/Caesium'
COBALT_NAI='NaI/Cobalt'
AMERICIUM_NAI='NaI/Americium'
BARIUM_NAI='NaI/Barium'
BACKGROUND_NAI='NaI/Background/background_1000s.Spe'

#csv file containing fitted peak data
NAI_PEAKS="Fitted_Data/NaI/total_nai.csv"

#region of interest for each peak
NAI_ROI={'cs':(500,700),
         'mid_ba':(55,110),
         'ba':(225,400),
         'co':(750,1000),
         'am':(0,85),
         'am1':(0,40),
         'am2':(40,85),
         'ka':(15,50)
         }

#date and time for original source actvitiy
UPSTAIRS_SOURCES=datetime.datetime(1979,2,1,12)
DOWNSTAIRS_SOURCES=datetime.datetime(1979,12,1,12)

Model = namedtuple('Model', ['name','model', 'estimates',
                             'params', 'components'])


#function parameters for each fit
GAUSSIAN_PARAMS = ('mu', 'sig', 'A')
GAUSSIAN_LINE_PARAMS = ('mu','sig','A','Slope','Intercept')
DOUBLE_LINE_PARAMS= ('mu1','sig1','A1',
                     'mu2','sig2','A2',
                     'Slope','Intercept')
DOUBLE_PEAK_PARAMS= ('mu1','sig1','A1',
                     'mu2','sig2','A2')


GAUSSIAN_MODEL = Model(
    name      = 'gaussian',
    model     = fit.gaussian,
    estimates = fit.gaussian_initial_estimates,
    params    = GAUSSIAN_PARAMS,
    components = fit.gaussian
)


GAUSSIAN_LINE_MODEL = Model(
    name      = 'gaussian line',
    model     = fit.gaussian_line,
    estimates = fit.gaussian_line_initial_estimates,
    params    = GAUSSIAN_LINE_PARAMS,
    components  = fit.gaussian_line_components
)

DOUBLE_GAUSSIAN_MODEL = Model(
    name      = 'double line',
    model     = fit.double_line,
    estimates = fit.double_line_initial_estimates,
    params    = DOUBLE_LINE_PARAMS,
    components = fit.double_gaussian_components
)


TWO_GAUSSIAN_MODEL = Model(
    name      = 'double peak',
    model     = fit.double_peak,
    estimates = fit.double_peak_initial_estimates,
    params    = DOUBLE_PEAK_PARAMS,
    components = fit.double_peak_components
)

#################################################

def nai_spectra():
    caesium_list=fit.plot_content(CAESIUM_NAI, BACKGROUND_NAI,DOWNSTAIRS_SOURCES)        
    barium_list=fit.plot_content(BARIUM_NAI, BACKGROUND_NAI, DOWNSTAIRS_SOURCES)
    americium_list=fit.plot_content(AMERICIUM_NAI, BACKGROUND_NAI, DOWNSTAIRS_SOURCES)    
    fit.everything(barium_list, NAI_ROI['mid_ba'],'mid_ba', GAUSSIAN_LINE_MODEL,'NaI')
    fit.everything(barium_list, NAI_ROI['ba'],'ba', DOUBLE_GAUSSIAN_MODEL,'NaI')
    fit.everything(caesium_list, NAI_ROI['ka'],'ka',  GAUSSIAN_LINE_MODEL,'NaI')
    fit.everything(caesium_list, NAI_ROI['cs'],'cs', GAUSSIAN_LINE_MODEL,'NaI')  
    fit.everything(americium_list, NAI_ROI['am2'],'am2', GAUSSIAN_LINE_MODEL,'NaI')
    
    
if __name__ == "__main__":
    nai_spectra()
    calib.total_calibration_spec(NAI_PEAKS)
    eff.final_source_efficiencies('NaI')    
```
